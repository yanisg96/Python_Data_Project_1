```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.describe()
```

```python
df.info()
```

## Useful Methods

df.count() - Count of values

df.sum() - Sum of values

df.cumsum() - Commulative sum of values

df.min()/df.max() - Minimum/Maximum values

df.idxmin()/df.idxmax() - Indexes of min/max value

df.mean() - Mean of values

df.median() - Median of values

df.mode() - Mode of the values

df.column_name.value_countrs() - Unique counts`

```python
df['job_title_short'].unique()
```

```python
# instead of using unique(), use .value_counts
df['job_title_short'].value_counts()
```

```python
# to find out what are the min salary avg of each of the job title short positions
# use .groupby, syntax is:
# df.groupby("columns to group")["columns to aggregate"].you_agg_method()

df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
# aggregating two different columns
# df.groupby("list of columns to group")["list of columns to aggregate"].you_agg_method()
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

```python
# use .agg to add multiple arguments like minimum/maximum, median, etc
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])
```
