## Logical Operators

and - x and y

or - x or y

not - not x

```python
job_work_from_home = False
job_health_insurance = False
job_no_degree_mention = True

job_work_from_home or job_health_insurance or not job_no_degree_mention
```

```python
not True
```

```python
if job_work_from_home or job_health_insurance or not job_no_degree_mention:
  print("meets the conditions")
```

## Membership Operators

in - Membership - x in y

not in - Not membership - x not in y

```python
# it's case sensitive (Aa)
"data" in "Data Nerd"
```

```python
job_skills = ['python', 'r', 'sql']
```

```python
'python' in job_skills
```

```python
'python' not in job_skills
```

## Identity Operators

is - Identity - x is y

is not - Not identity - x is not y

```python
# 'is' verifies if the two variables are using the same memory location for the
# value that they are holding

salary_kelly = 100_000
salary_luke = 100_000

salary_kelly is salary_luke
```

```python
#  if you were to use == instead, it tells you if they have the same value
salary_kelly == salary_luke
```

```python
id(salary_kelly)
```

```python
id(salary_luke)
```

```python
id(salary_kelly) == id(salary_luke)
```

```python
core_skills = ['Python', 'R', 'SQL']
lukes_skills = core_skills
```

```python
lukes_skills is core_skills
```
