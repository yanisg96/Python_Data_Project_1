```python
# Install datasets Library
!pip install datasets

# Import Libraries
import pandas as pd
from datasets import load_dataset

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)
```

```python
# .any() here is just to show us is there is (True) or not (False) otherwise we would get a list from each of the rows telling us True/False
df['job_country'].isin(['North Korea']).any()
```

```python
mx_jobs = df[df['job_country'] == 'Mexico']
```

```python
mx_jobs = mx_jobs[mx_jobs['salary_year_avg'].notna()]

mx_jobs
```
