```python
# Install datasets Library
!pip install datasets

# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Bar Chart - Counts of Job Posting

```python
job_counts = df.job_title_short.value_counts()
```

```python
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Counts of Job Posting')
plt.xticks(rotation=45, ha= 'right') # rotates the x axis by 45 degrees and ha is horizontal alignment
plt.show()
```

```python
plt.barh(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.xlabel('Counts of Job Posting')
plt.show()
```
