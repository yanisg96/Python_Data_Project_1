## Loops



*   **for** loops
*   **while** loops



```python
numbers = [0, 1, 2, 3, 4, 5]

# number is the variable, numbers is the iterable
for number in numbers:
  print(number)
```

```python
characters = 'PYTHON'

for character in characters:
  print(character)
```

```python
years_experience = {
    'Luke': 3,
    'Kelly': 6,
    'Ken': 4,
    'Alex': 3
}
```

```python
for person in years_experience:
  print(person)
```

```python
for value in years_experience.values():
  print(value)
```

```python
for key, value in years_experience.items():
  print(key, value)
```

```python
for key, value in years_experience.items():
  if value > 5:
    print(key, 'has', value, 'years of experience')
```

```python
count = 1
while count < 5:
  print(count)
  count += 1
```

```python
years_experience = {
    'Luke': 3,
    'Kelly': 6,
    'Ken': 4,
    'Alex': 3,
    'Mary': 7
}
```

```python
years = 0
index = 0
# a list of set items
years_list = list(years_experience.items())

while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, 'has', years, 'years of experience')
  index += 1
```
