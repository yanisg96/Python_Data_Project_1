## Sets

With sets we still use curly brackets {} but without the colon **":"** so that it doesn't turn into dictionaries.

Sets are unique, so if there's a duplicate, it will remove them.

There's also no index.

```python
job_skills = {'tableau', 'sql', 'python', 'statistics'}
```

```python
job_skills
```

```python
help(set)
```

```python
job_skills.add('looker')
```

```python
job_skills
```

```python
# if you try to add something that is already in the set, it won't duplicate it
job_skills.add('sql')
```

```python
job_skills
```

```python
# if we use .pop(), it will randomnly choose one of the values and remove them
job_skills.pop()
```

```python
# if we want to remove an specific value, we use .remove
```

```python
job_skills.remove('statistics')
```

```python
job_skills
```

```python
skill_list = ['python', 'sql', 'statistics', 'tableau', 'python', 'sql', 'statistics', 'tableau']
```

```python
# to turn the list into set, we do set(list)
set(skill_list)
```

```python
# and to turn it back to a list would be
list(set(skill_list))
```
