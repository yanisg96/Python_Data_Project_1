## String Formatting

*    Concatenation - 'string' + variable

*    str.format() - 'string{}'.format(variable)

*    f-strings - f'string {variable}'

*    % Formatting - 'string %s' % variable

*    join() - 'separator'.join(['Hello', name])


## Format method *.format()*:



args = arguments

kwargs = keyword arguments

```python
# format(...)
#  |      S.format(*args, **kwargs) -> str
#  |
#  |      Return a formatted version of S, using substitutions from args and kwargs.
#  |      The substitutions are identified by braces ('{' and '}')
# Create the string "Role: Data Analyst"
role = 'Data Analyst'
skill = 'Python'

"Role: {}".format(role)
```

**quick note:** to put # before an entire text that has been copied & pasted, you select the text then click **ctrl** + forward slash (**/**)

**About the format method**

what you put inside the **()** in the method **.format()** is going inside the **{}** in your string. ie. role **{}**

```python
# Create the string "Role: Data Analyst; Skill Required: Python"
"Role: {}; Skill Required: {}".format(role, skill)
```

```python
# We can also use the keywords like this
"Role: {role}: Skill Required: {skill}".format(role=role, skill=skill)
```

## f-Strings - Formatted String Literals

```python
role = 'Data Analyst'
skill = 'Python'

f"Role: {role}; Skill Required: {skill}"
```

## %-formatting (printf-style string formatting)

```python
"Role: %s" % (role)
```

```python
"Role: %s; Skill Required: %s" % (role,skill)
```
