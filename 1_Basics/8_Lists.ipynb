## Slicing

Syntax: **list[start:end:step]**

**start:** the starting index (inclusive)

**end:** the ending index (exclusive, meaning it doesn't mention the specified one)

**step:** steps to take between items, defaults to one

```python
job_skills = ['sql', 'excel','python']
```

```python
job_skills[0:3]
```

```python
# 0 and 3 are the default numbers of the index
# so using : still works the same as 0:3
job_skills[:]
```

```python
job_skills[:1]
```

```python
job_skills[::1]
```

```python
job_skills[::2]
```

```python
luke_skills = ['python', 'bigquery', 'r']
kelly_skills = ['python', 'sql', 'looker']

all_skills = luke_skills + kelly_skills

all_skills
```

```python
all_skills[::2]
```

```python
all_skills[5:]
```

```python
# the index of lists can be counted from left to right starting from 0
# but also from right to left starting on -1, then -2, -3 and so on
# so in this example, 5 and -1 are the same item

all_skills[-1:]
```

```python
all_skills[-2:]
```

## Unpacking

Assign each value in an iterable to a variable in a single statement

```python
job_skills = ['python', 'excel', 'sql','looker']
```

```python
skill1, skill2, skill3 = job_skills
```

```python
print(skill1)
print(skill2)
print(skill3)
```

```python
# the * is an unpacking operator,
# that assigns the rest of the items to be assigned
skill_concerned, *skill_dont_care = job_skills
```

```python
print(skill_concerned)
print(skill_dont_care)
```
