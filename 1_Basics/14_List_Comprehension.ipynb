## List Comprehension

Both are ways to do the same lists on number, but the first one only needs one line compared with the second.

```python
numbers = [x for x in range(100)]
```

```python
numbers = []

for number in range(100):
  numbers.append(number)
```

```python
numbers
```

```python
# you can multiply it by two with the *

numbers = [x*2 for x in range(100)]

numbers
```

```python
# but you can also convert x to a float
numbers = [float(x) for x in range(100)]

numbers
```

Not limited to only **list** comprehension:

*   set Comprehension
*   tuple Comprehension

```python
numbers = [x for x in 'PYTHON']
numbers
```
