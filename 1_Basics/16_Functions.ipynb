## User-Defined Functions

```python
base_salary = 100000
bonus_rate = 0.1

total_salary = base_salary * (1 + bonus_rate)

total_salary
```

```python
#Calculate total_salary of base_salary * (1 + bonus_rate)
# adding the bonus_rate=.1 in the function, makes it optional to put in the arguments

def calculate_salary(base_salary, bonus_rate=.1):

  total_salary = base_salary * (1 + bonus_rate)

  return total_salary
```

```python
salary_1 = 100000
rate_1 = 0.1

# so you can just use the salary_1 and no rate_1 since it's optional and it will
# use the one above in the function (bonus_rate=.1)

calculate_salary(salary_1)
```

```python
# but if you have another bonus rate, you can also put it in the function
salary_1 = 100000
rate_1 = 0.5

calculate_salary(salary_1,rate_1)
```
