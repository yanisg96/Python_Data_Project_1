## Dictionaries

The dictionary uses {} curly brackets, the first part is called **key** + **:** + **value**.

Example:

{'key1': 'value1', 'key2': 'value2'}

The keys can be int, strings, but not lists, however the values can be anything.

```python
# to show the type of values and keys that can be used
{'database': ['postgresql', 4, 'string', (1,2)]}
```

```python
job_type_skills = {
    'database': 'postgresql',
    'language': 'python',
    'library': 'pandas'
}
```

```python
job_type_skills
```

```python
# to call the key of database, use the []
job_type_skills['database']
```

```python
help(dict)
```

```python
# to call the keys inside the dictionary we use the .keys()
job_type_skills.keys()
```

```python
# and to call the values of the dictionary we can use .values()
job_type_skills.values()
```

```python
# to remove something from the dictionary, we use .pop(key)
job_type_skills.pop('library')
```

```python
job_type_skills
```

```python
# to add something to the dictionary, we can use .update()
job_type_skills.update({'cloud': 'google cloud'})
```

```python
job_type_skills
```

```python
# another easier way to update the dictionary would be to define first like this:
job_type_skills['version_control'] = 'git'
```

```python
job_type_skills
```
