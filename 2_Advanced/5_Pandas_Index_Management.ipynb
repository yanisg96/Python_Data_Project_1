```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index
```

```python
df.index.name = 'job_index'
```

```python
df.head()
```

```python
df.index.dtype
```

## Three Methods to cover
*  df.reset_index()
*  df.set_index()
*  df.sort_index()

## reset_index()

```python
df_mex = df[df['job_country'] == 'Mexico']
```

```python
df_mex.reset_index(inplace=True)

df_mex
```

## set_index()

```python
df_mex.set_index('job_index', inplace=True)
```

```python
df_mex
```

## sort_index()

```python
median_pivot = df_mex.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot
```

```python
median_pivot.sort_values(by= ('median', 'salary_year_avg'), ascending=False)
```

```python
# best use case = whenever the index doesn't have a name

median_pivot.sort_index(inplace=True)
```
