## Tuples

With tuples, you can't remove nor add more items. It is represented with ().

Tuples are for fixed data. Items cannot be changed.

Tuples are indexable.

```python
lukes_skills = ('python', 'sql', 'statistics', 'tableau')
lukes_skills
```

```python
# calling the first item, the index starts with 0
lukes_skills[0]
```

```python
# to call the first two, it would be like this
lukes_skills[0:2]
```

```python
# if we try to add any new items, we will get an error
lukes_skills.append('r')
```

```python
help(tuple)
```

```python
# with tuples
lukes_skills = ('python', 'sql', 'statistics', 'tableau')
print(id(lukes_skills))

new_skills = ('excel', 'power bi')

# if we wanted to add more skills, we could do the next
lukes_skills += new_skills

lukes_skills
id(lukes_skills)
```

```python
# with lists
lukes_skills = ['python', 'sql', 'statistics', 'tableau']
print(id(lukes_skills))

new_skills = ['excel', 'power bi']

lukes_skills += new_skills

lukes_skills
id(lukes_skills)
```
