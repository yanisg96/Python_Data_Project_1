```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
type(df.job_posted_date[0])
```

```python
# we want to convert str to datetime
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
# adding a new column called job_posted_month with only the month of the date value
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df
```

```python
# to sort the values ascendingly (default), and adding the inplace, it replaces df

df.sort_values(by='job_posted_date', inplace=True)
```

```python
df
```

```python
# to eliminate the column salary hour avg .drop(labels= column name, axis=0 if index or =1 if column)
df.drop(labels= 'salary_hour_avg', axis=1, inplace=True)
```

```python
df.info()
```

```python
# to drop/eliminate the rows that have NaN as values .dropna(axis=0 for rows, =1 for columns with the values)
# but in the case where we want to drop all the rows with NaN values we use subset=[name of the column]
df.dropna(subset=['salary_year_avg'], inplace=True)
```
