```python
import pandas as pd
from datasets import load_dataset

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# to only see the top rows of the table use .head(number of rows)
# if you leave () blank, it shows the first 5 values
df.head(10)
```

```python
# if you want to see the last rows use .tail(number of rows)
df.tail(10)
```

```python
# to call a specific columns use [] like a dictionary
df['job_title_short'].head()
```

```python
# and to call different columns, we add another pair of [] outside
df[['job_title_short', 'job_location']].head()
```

```python
#to show just certain row values and columns .iloc[90] would show just the 90th row
# and using .iloc[90:100] would show the range from row 90 to 100
# to see creatin row values and certain columns .iloc[90:100, 0:2]
df.iloc[90:100, 0:2]
```

```python
df[['job_title_short', 'job_title']].iloc[90:100]
```
