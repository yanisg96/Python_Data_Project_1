## Order of operations



1.   **Parentheses:** Anything in parentheses is executed first.
2.   **Exponents:** Operations involving exponents (including roots, indicated as fractional exponents).
3.   **Multiplication and Division:** Processed from left to right.
4.   **Addition and substraction:** Also processed from left to right.



```python
min_salary = 20_000 # _ is ignored by the interpreter, it's easier to read by us
max_salary = 80_000

avg_salary = (min_salary + max_salary) / 2

avg_salary
```

## Assignment Operators

Assign =

Add and assign +=

Substract and assign -=

Multiply and assign *=

Divide and assign /=

Modulus and assign %=

Exponent and assign **=

Floor divide and assign //=

```python
applicant_1 = "Kelly"
applicant_2 = "Luke"
```

The next one is one way to exchange the names of the applicants using a temporary variable

```python
temp_var = applicant_1
```

```python
applicant_1 = applicant_2
applicant_2 = temp_var
```

```python
applicant_1, applicant_2
```

An easier way with Python is an assignment within one line, which is the following

```python
applicant_1 = "Kelly"
applicant_2 = "Luke"
```

```python
applicant_1, applicant_2 = applicant_2, applicant_1
```

```python
applicant_1, applicant_2
```
